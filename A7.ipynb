```python
import diff_eqn as diff
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
#Q1
f1=lambda y,x:y*math.log(y)/x
SE1=diff.ode_forwardeuler(f1,2,2.71282,0.5,6)
SE2=diff.ode_forwardeuler(f1,2,2.71282,0.2,15)
SE3=diff.ode_forwardeuler(f1,2,2.71282,0.05,60)
x_=np.arange(2,5,0.01)
y_=np.exp(x_/2)
plt.plot(x_,y_,'--',label='Actual')
plt.plot(SE1[0],SE1[1],label='euler_h=0.5') 
plt.plot(SE2[0],SE2[1],label='euler_h=0.2') 
plt.plot(SE3[0],SE3[1],label='euler_h=0.05') 
plt.legend()
print("Here we have compared the solution obtained using euler method with the actual analytical solution")
```

```python
f1=lambda y,x:y*math.log(y)/x
SE1=diff.predCorr(f1,2,2.71282,0.5,6)
SE2=diff.predCorr(f1,2,2.71282,0.2,15)
SE3=diff.predCorr(f1,2,2.71282,0.05,60)
x_=np.arange(2,5,0.01)
y_=np.exp(x_/2)
plt.plot(x_,y_,'--',label='Actual')
plt.plot(SE1[0],SE1[1],label='predcor_h=0.5') 
plt.plot(SE2[0],SE2[1],label='predcor_h=0.2') 
plt.plot(SE3[0],SE3[1],label='predcor_h=0.05') 
plt.legend()
print("Here we have compared the solution obtained using predictor corrector method with the actual analytical solution")
```

```python
#Q2
F2=[lambda Y,x:Y[1],lambda Y,x: 1-x-Y[1]]
x_0=0
Y_0=[[2],[1]]
S=diff.RK4_solve_system(F2,x_0,Y_0,0.05)
x_0=0
Y_0=[[2],[1]]
S_=diff.RK4_solve_system(F2,x_0,Y_0,-0.05)
x=np.arange(-5,5,0.01)
y=1+np.exp(-x)-x**2/2+2*x
plt.plot(x,y,label='actual')
plt.plot(S[-1],S[0],label='RK_4')
plt.plot(S_[-1],S_[0],label='RK_4-')
plt.ylim(-5,5)
plt.xlim(-5,5)
plt.legend()
print('Here the analytical solution is completely identical with the RK4 solution')
```
